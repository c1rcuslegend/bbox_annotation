```python
# Inspiration and mobilenet.json file from: https://observablehq.com/@mbostock/imagenet-hierarchy
import json

data = json.load(open('mobilenet.json'))

def find_leaves(node, parents):
    if 'children' not in node:
        leaves.append((node['index'], node['name'], parents))
    else:
        for child in node['children']:
            find_leaves(child, parents + [node['name']])

leaves = []
find_leaves(data, [])

for leaf in leaves:
    print(f"Leaf: {leaf[1]}, Index: {leaf[0]}, Parents: {' -> '.join(leaf[2][1:])}")
```

```python
import json

def process_file(input_file):
    index_to_parent = {}
    parent_to_children = {}

    with open(input_file, 'r') as file:
        for line in file:
            if 'Leaf:' in line and 'Index:' in line and 'Parents:' in line:
                parts = line.split('Index:')
                leaf_name = parts[0].replace('Leaf:', '').strip().strip(',')

                index_parent = parts[1].split('Parents:')
                index = int(index_parent[0].strip().strip(','))
                parent = index_parent[1].strip()

                # First JSON: index -> parent
                index_to_parent[index] = parent

                # Second JSON: parent -> [children]
                if parent not in parent_to_children:
                    parent_to_children[parent] = []
                parent_to_children[parent].append(index)

    # Write the results to JSON files
    with open('index_to_parent.json', 'w') as f:
        json.dump(index_to_parent, f, indent=4)

    with open('parent_to_children.json', 'w') as f:
        json.dump(parent_to_children, f, indent=4)

# Usage
process_file('filtered_class_rel.txt')
```
